```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
%matplotlib inline
```

```python
data = pd.read_csv('data/temperature.csv')
data
```

```python
data=data.drop(["month"], axis=1) # removing month column
data
```

```python
data.index.name='Month' #naming index as month
data
```

```python
ax = plt.gca()
fig,ax = plt.subplots(figsize = [20,6])
ax.boxplot(data)
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(axis='both',which='major', length=9, direction="in")
ax.tick_params(axis='y',which='minor', length=4, direction="in")
ax.tick_params(axis='x', rotation= 45)
ax.set_xlabel('Month',fontsize =18,color='b')
ax.set_ylabel('Temperature($^o$C)',fontsize =18,color='b')
ax.set_xticks(np.arange(1,13))
ax.set_xticklabels(['Jan', 'Feb', 'March', 'April', 'May','Jun', 'July', 'Aug', 'Sept', 'Oct','Nov', 'Dec'])
plt.title('Monthly variation of temperature',fontsize =24,color='r')
fig.tight_layout() 
plt.savefig("image/box1.pdf", dpi = 300) # dpi dot per inch
plt.show()
```

```python
ax = plt.gca()
fig,ax = plt.subplots(figsize = [20,6])
ax.boxplot(data.T)
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(axis='both',which='major', length=9, direction="in")
ax.tick_params(axis='y',which='minor', length=4, direction="in")
ax.tick_params(axis='x', rotation= 45)
ax.set_xlabel('Year',fontsize =18,color='b')
ax.set_ylabel('Temperature($^o$C)',fontsize =18,color='b')
ax.set_xticks(np.arange(1,11))
ax.set_xticklabels(['2001', '2002', '2003', '2004', '2005','2006', '2007', '2008', '2009', '2010'])
plt.title('Annual variation of temperature',fontsize =24,color='r')
fig.tight_layout() 
plt.savefig("image/box2.pdf", dpi = 300) # dpi dot per inch
plt.show()
```

```python
myear= np.mean(data,axis =0)
print(myear)
```

```python
mmonth=np.mean(data,axis =1)
print(mmonth)
```

```python
sdyear=np.std(data,axis =0)
print(sdyear)
```

```python
sdmonth=np.std(data,axis = 1)
print(sdmonth)
```

```python
plt.figure(figsize = [20,6])
fig,ax = plt.subplots()
plt.errorbar(data.index, mmonth,yerr=sdmonth,fmt='-o')
plt.xlabel('Month',fontsize =14,color='b')
plt.ylabel('Temperature($^o$C)',fontsize =18,color='b')
plt.xticks(data.index, ('Jan', 'Feb', 'March', 'April', 'May','Jun', 'July', 'Aug', 'Sept', 'Oct','Nov', 'Dec'),fontsize =18)
plt.xticks(rotation=45)
plt.title('Monthly variation of temperature ',fontsize =24,color='g')
plt.tight_layout() 
plt.savefig("image/errorbar1.pdf", dpi = 300) # dpi dot per inch
plt.show()
```

```python
plt.figure(figsize = [20,6])
fig,ax = plt.subplots()
plt.bar(data.columns, myear,yerr=sdyear,align='center', color="c", ecolor='black')
plt.xlabel('Year',fontsize =14,color='b')
plt.xticks(rotation=45)
plt.ylabel('Temperature($^o$C)',fontsize =14,color='b')
plt.title('Annual variation of temperature',fontsize =18,color='g')
plt.tight_layout() 
plt.savefig("image/errorbar2.pdf", dpi = 300) # dpi dot per inch
plt.show()
```
